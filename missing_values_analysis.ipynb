```python
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
#reading the dataset
df = pd.read_csv('https://raw.githubusercontent.com/FazlyRabbiBD/AcademicWorkspace/master/survey-data-missing-values.csv')
```

```python
df
```

```python
#9
#Renaming Columns with inplace
df.rename(columns = {"EducationLevel": "HighestEducationalDegree"},inplace=True) 
```

```python
df.head()
```

```python
#Check Null
df.isnull().values.any()
```

```python
#10
#Count the number of Null values
df.isnull().values.sum()
```

```python
#11
#Showing the null values as per attributes
null_columns=df.columns[df.isnull().any()]
df[null_columns].isnull().sum()
```

```python
#12
df['AnnualIncome'] = pd.to_numeric(df['AnnualIncome'], errors='coerce')
df
```

```python
#13
# Filling Null with group by vparameter
df['AnnualIncome'] = df['AnnualIncome'].fillna(df['AnnualIncome'].median())
df
```

```python
#14
df['District']=df['District'].fillna(method='ffill')
df
```

```python
#15
df['Age'] = pd.to_numeric(df['Age'], errors='coerce')
df
```

```python
#16
# Filling Null with group by vparameter
df['Age'].fillna(df['Age'].mode()[0], inplace=True)
df
```

```python
#17
df['MemberOfHousehold'].fillna(df['MemberOfHousehold'].mode()[0], inplace=True)
df
```

```python
#18
#unwanted value treatment
cnt=0
for row in df['Occupation']:
    try:
        int(row)
        df.loc[cnt, 'Occupation']=np.nan
    except ValueError:
        pass
    cnt+=1
df
```

```python
#19

df['Occupation'].fillna(df['Occupation'].mode()[0], inplace=True)
df
```

```python
#20
df['Gender'].fillna(df['Gender'].mode()[0], inplace=True)
df
```

```python
#21
#Converting street number to int
df.Age = df.Age.astype('int64')
df.MemberOfHousehold = df.MemberOfHousehold.astype('int64') 
df.AnnualIncome = df.AnnualIncome.astype('int64') 
df
```

```python
#22
from sklearn.preprocessing import LabelEncoder 
le = LabelEncoder() 
df['Occupation']= le.fit_transform(df['Occupation'])
df['HeadOfTheHousehold']= le.fit_transform(df['HeadOfTheHousehold'])
df['HighestEducationalDegree']= le.fit_transform(df['HighestEducationalDegree']) 
df
```

```python
#23
gender = {'Male' :1 , 'Female' : 2 }
df['Gender'] = df['Gender'].map(gender)

district = {'Bagerhat' :1 , 'Patuakhali' : 3, 'Khulna': 3 }
df['District'] = df['District'].map(district)
df
```

```python
#24
from sklearn.preprocessing import StandardScaler
std = StandardScaler()
df_scaledstd = pd.DataFrame(std.fit_transform(df),columns = df.columns)
df_scaledstd
```
